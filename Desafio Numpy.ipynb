# Desafio: Usando matrizes RGB (imagens) no Numpy

Toda imagem colorida, para o computador, nada mais é do que um conjunto de três matrizes (vermelha, verde e azul (RGB de red, green e blue)), onde cada uma dessas matriz traz a intensidade da respectiva cor para cada um dos pixels.
![image.png](imgs/img5.jpg)

Os valores vão de 0 a 255, onde 0 representa a ausência dessa cor e 255 a intensidade mais forte. Se tivermos um píxel puramente vermelho ele será representado pelos seguintes valores em cada matriz:

- R = 255
- G = 0
- B = 0

Na imagem acima, cada célula da matriz representa um píxel. Uma imagem é formada por vários pixels.

Já uma imagem preta e branca ou também chamada de escala de cinza apresenta apenas uma matriz com valores entre 0 e 255 ou entre 0 e 1, onde quanto mais perto de 0, mais escuro é o píxel e quanto mais distante, mais claro.

![image-2.png](imgs/img6.jpg)

## Carregando a imagem e visualizando

````python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
````

````python
img = io.imread('shiba-p.jpg') # Carregamento da imagem

print(type(img))

plt.imshow(img) # Exibição da imagem
````

````python
img.shape 
````

````python
img_copia = img.copy()
plt.imshow(img_copia)
````

## 1. Cortando no meio vertical

````python
img_copia1 = img.copy()
img_copia1 = img_copia1[:,:-320]
plt.imshow(img_copia1)
````

## 2. Cortando no meio horizontal

````python
img_copia2 = img.copy()
img_copia2 = img_copia2[:-320, :]
plt.imshow(img_copia2)
````

## 3. Apagando o nariz

````python
img_copia3 = img.copy()
img_copia3[280:380,320:420] = 255
plt.imshow(img_copia3)
````

## 4. Transformando para escala de cinza

A fórmula para transforma para escala de cinza é:

$$0.299 * \text{red} + 0.587 * \text{green} + 0.114 * \text{blue}$$

E para exibir a matriz gerada use:

```python
plt.imshow(nome_da_sua_matriz, cmap="gray")
```

Obs: Lembre-se que para a escala de cinza o resultado deve ser apenas uma matriz, ou seja, o shape será (640, 640)

````python
image_copia4 = img.copy()
gray_image = image_copia4[:,:,0] * 0.299 + image_copia4[:,:,1] * 0.587 + image_copia4[:,:,2] * 0.114
plt.imshow(gray_image, cmap='gray')
````
